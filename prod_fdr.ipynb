```python
import datetime as dt
from etl_tool import FdrBroker, FdrExtractor, FdrOhlcvPreproc
```

```python
fdr_broker =  FdrBroker()
fdr_extractor = FdrExtractor(fdr_broker)
fdr_extractor
```

```python
krx_df = fdr_extractor.extract_krx_df()
krx_df
```

```python
kospi_df = krx_df[krx_df['Market'] == 'KOSPI']
kosdaq_df = krx_df[krx_df['Market'] == 'KOSDAQ']
```

```python
kospi_stock_codes =  list(set(kospi_df['Code']))
kosdaq_stock_codes =  list(set(kosdaq_df['Code']))
```

```python
end_date = dt.date.today()
start_date = end_date - dt.timedelta(days=365 * 5)
```

```python
kospi_ohlcv_df = fdr_extractor.extract_multi_ohlcv_df(
    kospi_stock_codes, start_date, end_date
)
```

```python
kosdaq_ohlcv_df = fdr_extractor.extract_multi_ohlcv_df(
    kosdaq_stock_codes, start_date, end_date
)
```

```python
fdr_ohlcv_preproc = FdrOhlcvPreproc()
```

```python
kospi_ohlcv_df_ = fdr_ohlcv_preproc(kospi_ohlcv_df)
kosdaq_ohlcv_df_ = fdr_ohlcv_preproc(kosdaq_ohlcv_df)
```

```python
kospi_ohlcv_df_.to_csv("./data/fdr/kospi_ohlcv_df.csv")
kosdaq_ohlcv_df_.to_csv("./data/fdr/kosdaq_ohlcv_df.csv")
```
